```python
## Importing libraries
import pandas as pd
import numpy as np
import scipy as sci
import seaborn as sns
import matplotlib.pyplot as plt

## Import data about global terrorism from https://www.kaggle.com/START-UMD/gtd

data = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1')
data.shape
data = data.loc[:,['eventid','iyear','imonth','iday','country_txt','region_txt','city','latitude','longitude',
                    'attacktype1_txt','targtype1_txt','success','gname','weaptype1_txt',
                    'nkill','nwound']]
## Question no. 1 What percent of terrorist attacks are successful? 
all_attacks = len(data)
#print(all_attacks)
success = data[data.success == 1]
success=len(success)
#print(success)
print(" {a} total number of attacks - including {p:.2f}% of successful attacks".format(a=all_attacks, p=success/all_attacks*100))
```

```python
## Question no. 2 What is the number of terrorist attacks per year? 
barplot = pd.value_counts(data['iyear'])\
.sort_index()\
.plot\
.bar(width=0.8, figsize=(15, 7), title="Amount of terrorist attacks per year")
plt.show()
```

```python
## Question no. 3 Amount of terrorist attacks per region

count_region = data.groupby(['region_txt']).count()
plt.figure(figsize = (15,5))
plt.bar(count_region.index, count_region.eventid)
plt.xlabel('Region')
plt.xticks(rotation =45)
plt.show()
```

```python
## Question no. 4 What kind of terrorist attack is the most popular? 
ax = sns.countplot(y="attacktype1_txt", data=data)
ax.set_xlabel("Amount of attacks")
ax.set_ylabel("Attack type")
plt.show()
```

```python
## Question no. 5 What percentage of terrorist attacks has been occurred in United States? 
USA = data[data.country_txt == 'United States']
United_States=len(USA)

print("{p:.2f}% of all terrorist attacks in the whole worldwide has been occurred in USA".format(a=all_attacks, p=United_States/all_attacks*100))
```
